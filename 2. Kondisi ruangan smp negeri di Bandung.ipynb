## Clustering berdasarkan Kondisi Ruangan

```python
df2=pd.DataFrame()
df2['Baik']=df['Baik']/df["Jumlah Ruangan"]
df2['Rusak Ringan']=df['Rusak Ringan']/df["Jumlah Ruangan"]
df2['Rusak Sedang']=df['Rusak Sedang']/df["Jumlah Ruangan"]
df2['Rusak Berat']=df['Rusak Berat']/df["Jumlah Ruangan"]
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X=df2
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=100,n_init=1)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
#plt.savefig('elbow.png') #save png
plt.show()
```

```python
import numpy as np
X = df2
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=100,n_init=1)
kmeans.fit(X)
kmeans.labels_
```

```python
df["Cluster"]=kmeans.labels_
df.sort_values(by="Cluster")
```

```python
df["Cluster"]=df["Cluster"].replace([0,1,2],["Sedang","Baik","Kurang"])
df.head()
```

```python
df["Cluster"].value_counts()
```

```python
plt.figure(figsize=(4,4))
ax = df["Cluster"].value_counts()
ax = ax.plot(kind='bar', title="Cluster Sekolah berdasarkan Kondisi Ruangan Kelas", fontsize=14)
```

Saya tertarik untuk menggelompokkan SMP Negeri di Bandung berdasarkan kondisi ruang kelasnya. Pada pengolahan ini, saya ubah data per kondisi kedalam prosentase. Sebab jumlah ruang kelas tiap sekolah berbeda beda. Pada pengelompokkan ini, saya mempergunakan kmeans++ dengan membaginya menjadi 3 kelas. Sebenarnya bisa dibagi 2, hanya saja kurang 'asyik'. Setelah diperoleh hasil 'Clustering'nya saya ubah nama cluster berdasarkan karakteristik tiap kluster. Diperoleh bahwa terdapat 16 sekolah kondisi Baik, 33 sekolah kondisi Sedang dan 8 sekolah kondisi Kurang. Tentunya 8 sekolah inilah yang perlu menjadi perhatian utama untuk perbaikan ruangan kelas.
